#Q1

```python
import numpy as np
X=np.random.randint(1,51, size=(100,10))
# print(X)
#finding the mean,median, std, max of the entire array
print("1. The mean of the array is: ", X.mean())
print("1. The median of the array is: ", np.median(X))
print("1. The standard devation of the array is: ", X.std())
print("1. The maximum number in the array is: ", np.max(X))
print("-------------------------------------------------------------")
print("The mean,median,standard deviation, and the max of each row in the array is")
print()

#finding mean,median, std, max of each row in the array
for x in X: #or for x in range(100)
  print("The mean of", x, "row is: ", X[x].mean())
  print("The median of", x, "row is: ", np.median(X[x]))
  print("The standard deviation of", x, "row is: ", X[x].std())
  print("The maximum of", x, "row is: ", X[x].max())
  print("-----------------------------------------------------------")
```

#Q2

```python
import numpy as np
Y=np.random.randint(1,51, size=(1000,))

#finding the first and last 10 elements in the array along with the variance
print("The original array is: ", Y)
print("The first 10 elements are ", Y[0:10])
print("The last 10 elements are ", Y[-10:])
print("The variance of the array is: ", Y.var())
print()

#finding the differnce between each adjacent element
difference=np.diff(Y)
print("The difference between adjacent elements is ", np.diff(Y))

#finding the sum of the difference array in two different methods
#1. converting it to a list and using the sum method
#2. using a for loop method
diff_list=[]
for x in difference:
  diff_list.append(x)
print("The sum of the above difference array is(using sum method): ",sum(diff_list))

#method-2
sum_ele=0
for x in difference:
  sum_ele+=x
print("The sum of the above difference array is alternative way(using for loop): ",sum_ele)
print()

#sorting the array
sorted_arr=np.sort(Y)
print("The array after sorting is ", np.sort(Y))

#finding the first and last 10 elements in the array along with the variance after sorting
print("The first 10 elements after sorting are ", sorted_arr[0:10])
print("The last 10 elements after sorting are ", sorted_arr[-10:])
print("The variance of the array after sorting is: ", sorted_arr.var())
print("The variance remained the same before and after sorting the array")
print()

#finding the mean of first, last 10 elements and the mean of the difference and the sum of elements
print("The mean of first 10 elements is: ", Y[0:10].mean())
print("The mean of last 10 elements is: ", Y[-10:].mean())
print("The mean of the difference elements is: ", difference.mean())
print("The mean of the sum of difference elements is: ", sum_ele.mean())
print("From the above data we can conclude that the mean of first and last 10 elements is near to 25 which implies that half of the data is scattered before 25 and after 25 as the range os only from 0-25")

#finding the mean, median, std, max of each column in the array, as this is an array of size (1000,1) the mean, median, max will be the number itself
#the std will be zero
for x in range(1,10): #the range is supposed to be (1,1000)
  print("The mean of", x, "column is: ", sorted_arr[:x].mean())
  print("The median of", x, "row is: ", np.median(sorted_arr[:x]))
  print("The standard deviation of", x, "row is: ", sorted_arr[:x].std())
  print("The maximum of", x, "row is: ", sorted_arr[:x].max())
  print("-----------------------------------------------------------")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#opening and reading the text file and converting it into a csv format
pop_array=np.array([])
pop=pd.read_csv("US_population.txt")
pop.to_csv("US_population.csv")

#accessing population and removing the dates from the list
population= pop[" Population"]
# print(population)

#converting the population data into a numpy array 
pop_array=np.append(pop_array, population)
# print(pop_array)

#finding the annual growth of the array, subtracting b-a
diff_array =  (pop_array[1:] - pop_array[0:len(pop_array)-1])
final_annual_growth=[]

#fining the annual growth of the population by diving with a i.e. b-a/a
for x in range(150):
  final_annual_growth.append(diff_array[x]/population[x])
print("The annula growth of the US population is from 1950-2100")
print(final_annual_growth)

#graphing the population from 1950-2100
#using solid line for historical data (1950-2020)
#using dashed line for UN projection(2020-2100)
#population[70] is where the division of population before 2020 and after 2020 is divided, so population[71] is used as a seperater
plt.plot(population[:71], linestyle="solid")
plt.plot(population[71:], linestyle="dashed")

plt.plot(population[:71], 'b-', label='Historical Data')
plt.plot(population[71:], label='Un Projections', ls="--")
plt.legend(loc='upper left')
plt.show()

#graphing the annual growth of the array in the same way as he population is being plotted
plt.plot(np.array(final_annual_growth[:71]), linestyle="solid")
plt.plot(np.array(final_annual_growth[71:]), linestyle="dashed")

plt.plot(np.array(final_annual_growth[:71]), 'b-', label='Historical Data')
plt.plot(np.array(final_annual_growth[71:]), label='Un Projections', ls="--")
plt.legend(loc='upper right')
plt.show()
```
