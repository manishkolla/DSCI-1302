#NBA Stats

```python
'''NBA Data Set: https://www.kaggle.com/datasets/umutalpaydn/nba-20202021-season-player-stats?select=nba2021_advanced.csv'''

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
nba_read=pd.read_csv("nba2021_advanced.csv")
nba_read=nba_read.head(20)
# nba_read=nba_read.sample(10)
players=nba_read["Player"]
x=np.array(players)
ts=nba_read["TS%"]
per=nba_read["PER"]

print("Q1")
print("The player with maximum TS% value is", "Jared Dudley with a score of: ", max(ts))
print("The player with maximim PER value is ", "Giannis Antetokounmpo with a score of: ", max(per))
print()
print("Using 20 rows to plot the player efficiency and true shooting percentage")
plt.bar(x,nba_read["PER"], color="red", alpha=0.2)
plt.xticks(rotation="vertical")
plt.show()
print("True Shooting")
plt.scatter(nba_read["TS%"],x, color="blue", alpha=0.2)
plt.show()

print()
print("Q2")
q2=nba_read.sort_values(by="PER",ascending=False)
print("The top 10 players who player based on their performace this year are")
print(q2["Player"].head(10))
print()

print("Q3")
print("The worst 10 players who player based on their performace this year are")
print(q2["Player"].tail(10))
print()

print("Q4")
print("plotting overall data")
plt.plot(x,nba_read["PER"], color="red", alpha=1)
plt.plot(x,nba_read["TS%"], color="blue", alpha=1)
plt.xticks(rotation="vertical")
plt.show()
```
