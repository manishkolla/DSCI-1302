```python
import matplotlib.pyplot as plt
import numpy as np


x=np.arange(0,6*np.pi,0.2)
# x = np.linspace(-2*np.pi,2*np.pi,1000)

y=2*np.sin(x)
z=2*np.cos(x)
list=[]
for a in range(-4,5):
  list.append(a)

plt.plot(x,y, linestyle="solid")
plt.plot(x,z, linestyle="dashed")

# plt.xticks([-10,-7.5,-5,-2.5,0,2.5,5,7.5,10])
# plt.yticks(list)

# plt.xlim(-10,10,2.5)
plt.ylim(-4,4)

plt.xlabel("X")
plt.ylabel("Y")

plt.grid(color="blue",alpha=0.2)

plt.plot(x,y, 'b-', label='2sin(x)')
plt.plot(x,z, label='2cos(x)', ls="--")

plt.legend(loc='upper right')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.random.rand(1000) * 100
Y = np.random.normal(50, 10, 1000)
#X: generates a random 1000 number in an array and multiples with 100
#Y: the normal function is used to get a random array with a range and standard deviation given by the user
#the first number is the mean, second one is the standard deviation, and the third one is the size of the array(i.e. number of elements in the array )
print("The mean of X is: ", X.mean())
print("The mean of Y is: ", Y.mean())
print("The standard deviation of X is: ", X.std())
print("The standard deviation of Y is: ", Y.std())
print()
plt.hist(X, bins=20, alpha=0.6)
plt.hist(Y, bins=20, alpha=0.4)
plt.show()

#the frequency of the histogram is fluctuating between 40 and 60 in Y axis and the second histogram is increasing in a timely manner in between 20 and 80 on X axis
#both the histograms overlap in between 20 and 80 on X axis
```
